* Para asegurar que el modelo nos devuelva el nombre del mes en castellano, previamnte definimos el diccionario de meses en español.

```python
# Diccionario de traducción de meses
meses_es = {
    1: 'enero', 2: 'febrero', 3: 'marzo', 4: 'abril',
    5: 'mayo', 6: 'junio', 7: 'julio', 8: 'agosto',
    9: 'septiembre', 10: 'octubre', 11: 'noviembre', 12: 'diciembre'
}
```

**Crear titulares automáticos**

```python
# Crear titulares automáticos por día
def generar_titular_diario(row):
    dia = row['fecha'].day
    mes = meses_es[row['fecha'].month]
    anio = row['fecha'].year
    fecha_str = f"{dia:02d} de {mes} de {anio}"

    titular = f"El {fecha_str}, "
    frases = []

    # Temperatura máxima destacada
    if pd.notna(row['temperatura_maxima']) and row['temperatura_maxima'] >= 30:
        frases.append(f"ha llegado el calor a Madrid, alcanzando los {row['temperatura_maxima']}°C")
    elif pd.notna(row['temperatura_maxima']) and row['temperatura_maxima'] <= 10:
        frases.append(f"se ha registrado una temperatura máxima baja de {row['temperatura_maxima']}°C")
    elif pd.notna(row['temperatura_maxima']):
        frases.append(f"la temperatura máxima ha sido de {row['temperatura_maxima']}°C")

    # Temperatura mínima
    if pd.notna(row['temperatura_minima']) and row['temperatura_minima'] <= 5:
        frases.append(f"las noches han sido frías, bajando a {row['temperatura_minima']}°C")

    # Viento fuerte
    if pd.notna(row['racha_viento']) and row['racha_viento'] >= 50:
        frases.append(f"se han registrado rachas de viento fuertes de hasta {row['racha_viento']} km/h")

    # Precipitación
    if pd.notna(row['precipitacion']) and row['precipitacion'] > 0:
        frases.append(f"ha llovido un total de {row['precipitacion']} mm")

    # Si no hay eventos relevantes
    if not frases:
        frases.append("no se han registrado fenómenos destacables")

    titular += ", ".join(frases) + "."
    return titular

```

```python
df['titular'] = df.apply(generar_titular_diario, axis=1)
df[['fecha', 'titular']].head(10)
```

+ Utilizaremos monthly_stats, que ya agrupa por mes los datos.
Nos aseguramos de que se ha generado este DataFrame previamente:

```python
monthly_stats = df.groupby(['anio', 'mes']).agg({
    'temperatura_media': 'mean',
    'temperatura_maxima': 'max',
    'temperatura_minima': 'min',
    'precipitacion': 'sum',
    'duracion_sol': 'sum'
}).reset_index()

monthly_stats = monthly_stats.round(1)
```

**Crear noticias breves variadas por mes:**

```python
# Traducir número de mes a nombre en español
meses_es = {
    1: 'enero', 2: 'febrero', 3: 'marzo', 4: 'abril',
    5: 'mayo', 6: 'junio', 7: 'julio', 8: 'agosto',
    9: 'septiembre', 10: 'octubre', 11: 'noviembre', 12: 'diciembre'
}


import random

def generar_noticia(row):
    mes = meses_es[row['mes']]
    anio = int(row['anio'])

    frases = []

    # Introducción base
    inicio = random.choice([
        f"Durante el mes de {mes} de {anio},",
        f"En {mes} de {anio},",
        f"{mes.capitalize()} de {anio} ha estado marcado por"
    ])

    # Temperatura media
    if pd.notna(row['temperatura_media']):
        frases.append(f"una temperatura media de {row['temperatura_media']}°C")

    # Temperatura máxima destacable
    if pd.notna(row['temperatura_maxima']):
        frases.append(f"una máxima registrada de {row['temperatura_maxima']}°C")

    # Temperatura mínima
    if pd.notna(row['temperatura_minima']):
        frases.append(f"y una mínima de {row['temperatura_minima']}°C")

    # Precipitaciones
    if pd.notna(row['precipitacion']) and row['precipitacion'] > 0:
        frases.append(f"con un acumulado de {row['precipitacion']} mm de precipitaciones")
    else:
        frases.append("sin precipitaciones significativas")

    # Horas de sol
    if pd.notna(row['duracion_sol']) and row['duracion_sol'] > 0:
        frases.append(f"y un total de {int(row['duracion_sol'])} minutos de sol")

    # Construir el texto final
    cuerpo = ", ".join(frases)
    texto = f"{inicio} {cuerpo}."

    return texto
```

+ Aplicar la función

```python
monthly_stats['noticia_breve'] = monthly_stats.apply(generar_noticia, axis=1)

# Mostrar algunas noticias generadas
monthly_stats[['anio', 'mes', 'noticia_breve']].head(10)
```

Creamos y guardamos los resultados en un documento word:

```python
from docx import Document

# Crear un nuevo documento Word
doc = Document()
doc.add_heading('Resumen mensual del tiempo en Madrid', 0)

# Añadir cada noticia como párrafo
for index, row in monthly_stats.iterrows():
    noticia = row['noticia_breve']
    if pd.notna(noticia):
        doc.add_paragraph(noticia)

# Guardar el documento
nombre_archivo = 'noticias_meteorologicas_mensuales_madrid.docx'
doc.save(nombre_archivo)

from google.colab import files
files.download(nombre_archivo)
```

```python
import random

def generar_noticia(row):
    meses_es = {
        1: 'enero', 2: 'febrero', 3: 'marzo', 4: 'abril',
        5: 'mayo', 6: 'junio', 7: 'julio', 8: 'agosto',
        9: 'septiembre', 10: 'octubre', 11: 'noviembre', 12: 'diciembre'
    }

    mes = meses_es.get(row['mes'], 'mes desconocido')
    anio = int(row['anio'])

    # Introducciones compatibles con redacción posterior
    intro = random.choice([
        f"Durante el mes de {mes} de {anio},",
        f"En {mes} de {anio},",
        f"A lo largo de {mes} de {anio},"
    ])

    frases = []

    # Temperatura media
    if pd.notna(row['temperatura_media']):
        frases.append(f"una temperatura media de {row['temperatura_media']}°C")

    # Temperatura máxima
    if pd.notna(row['temperatura_maxima']):
        frases.append(f"una máxima que llegó a los {row['temperatura_maxima']}°C")

    # Temperatura mínima
    if pd.notna(row['temperatura_minima']):
        frases.append(f"y una mínima que descendió hasta los {row['temperatura_minima']}°C")

    bloque_temperatura = ", ".join(frases)

    # Precipitaciones
    if pd.notna(row['precipitacion']):
        if row['precipitacion'] == 0:
            bloque_lluvia = "No se registraron precipitaciones durante este mes"
        elif row['precipitacion'] < 10:
            bloque_lluvia = f"Las lluvias fueron escasas, con apenas {row['precipitacion']} mm acumulados"
        else:
            bloque_lluvia = f"Se registraron {row['precipitacion']} mm de lluvia en total"
    else:
        bloque_lluvia = ""

    # Horas de sol
    if pd.notna(row['duracion_sol']) and row['duracion_sol'] > 0:
        bloque_sol = f"El sol brilló durante {int(row['duracion_sol'])} minutos en total"
    else:
        bloque_sol = ""

    # Construcción final del texto
    noticia = f"{intro} Madrid presentó {bloque_temperatura}."
    if bloque_lluvia:
        noticia += f" {bloque_lluvia}."
    if bloque_sol:
        noticia += f" {bloque_sol}."

    return noticia.strip()
```

```python
import random

# Diccionario de meses en español
meses_es = {
    1: 'enero', 2: 'febrero', 3: 'marzo', 4: 'abril',
    5: 'mayo', 6: 'junio', 7: 'julio', 8: 'agosto',
    9: 'septiembre', 10: 'octubre', 11: 'noviembre', 12: 'diciembre'
}

def generar_noticia_breve_para_fecha(fecha, df):
    row = df[df['fecha'] == fecha]
    if row.empty:
        return f"No hay datos disponibles para {fecha.date()}."

    row = row.iloc[0]
    dia = fecha.day
    mes = meses_es[fecha.month]
    anio = fecha.year
    fecha_str = f"{dia:02d} de {mes} de {anio}"

    # Introducción
    intro = random.choice([
        f"El {fecha_str}, Madrid ha registrado",
        f"Durante el {fecha_str}, se observó en Madrid",
        f"A lo largo del {fecha_str}, el tiempo en Madrid se caracterizó por"
    ])

    frases = []

    # Temperatura media
    if pd.notna(row['temperatura_media']):
        frases.append(f"una temperatura media de {row['temperatura_media']}°C")

    # Temperatura máxima
    if pd.notna(row['temperatura_maxima']):
        frases.append(f"una máxima de {row['temperatura_maxima']}°C")

    # Temperatura mínima
    if pd.notna(row['temperatura_minima']):
        frases.append(f"y una mínima de {row['temperatura_minima']}°C")

    # Lluvia
    if pd.notna(row['precipitacion']):
        if row['precipitacion'] == 0:
            frases.append("sin precipitaciones registradas")
        elif row['precipitacion'] < 5:
            frases.append(f"con lluvias leves de {row['precipitacion']} mm")
        else:
            frases.append(f"con un total de {row['precipitacion']} mm de lluvia")

    # Sol
    if pd.notna(row['duracion_sol']) and row['duracion_sol'] > 0:
        frases.append(f"y {int(row['duracion_sol'])} minutos de sol")

    # Si no hay nada relevante
    if not frases:
        frases.append("condiciones meteorológicas normales sin eventos destacados")

    # Construcción final
    cuerpo = ", ".join(frases)
    noticia = f"{intro} {cuerpo}."

    return noticia.strip()
```

##VISUALIZACIÓN GENERADOS NOTICIAS STREAMLIT CLOUD


```python
import streamlit as st
import pandas as pd
from docx import Document
from io import BytesIO

# Funciones auxiliares que debes definir (o importar si ya las tienes)
def generar_titular_diario(row):
    return f"Titular automático para el {row['fecha'].strftime('%d-%m-%Y')}"

def generar_noticia_breve_para_fecha(fecha, df):
    return f"Resumen breve del tiempo para el {fecha.strftime('%d-%m-%Y')}."

def generar_noticia_con_comparativas(fecha_str, df):
    return f"Comparativa del tiempo con otros días en base al {fecha_str}."

# Interfaz Streamlit
st.title("📰 Generador de Noticias del Tiempo")

# Subida del archivo Excel
archivo = st.file_uploader("Sube el archivo Excel con datos del tiempo", type=["xls", "xlsx"])
if archivo:
    df = pd.read_excel(archivo)
    # Convertimos la columna 'fecha' a datetime si no lo está
    if not pd.api.types.is_datetime64_any_dtype(df['fecha']):
        df['fecha'] = pd.to_datetime(df['fecha'])

    # Selección de fecha
    fecha_input = st.date_input("Selecciona una fecha para generar la noticia")

    # Botón para generar
    if st.button("Generar Noticia"):
        try:
            # Buscar la fila
            row = df[df['fecha'] == pd.to_datetime(fecha_input)].iloc[0]

            # Generar textos
            titular = generar_titular_diario(row)
            noticia_breve = generar_noticia_breve_para_fecha(fecha_input, df)
            noticia_comparativa = generar_noticia_con_comparativas(fecha_input.strftime('%d-%m-%Y'), df)

            # Crear documento
            doc = Document()
            doc.add_heading(titular, level=1)
            doc.add_paragraph("")
            p_breve = doc.add_paragraph(style='List Bullet')
            run_breve = p_breve.add_run(noticia_breve)
            run_breve.bold = True
            doc.add_paragraph("")
            doc.add_paragraph(noticia_comparativa)

            # Guardar en memoria para descarga
            buffer = BytesIO()
            doc.save(buffer)
            buffer.seek(0)
            nombre_archivo = f"noticia_tiempo_{fecha_input.strftime('%Y_%m_%d')}.docx"

            st.success("✅ Noticia generada correctamente.")
            st.download_button("📥 Descargar documento Word", buffer, file_name=nombre_archivo, mime="application/vnd.openxmlformats-officedocument.wordprocessingml.document")
        except IndexError:
            st.error("❌ No se encontró información para la fecha seleccionada.")
```
